### Linearization using Taylor series:
#### f(x) ~ f(a) + f'(a)(x-a)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
class Linearizer:
    def __init__(self, f, df, init_x, a):
        self.x = init_x
        self.f_of_x = self.f_of_a(f, self.x)
        self.lin_of_x_at_a = self.t_of_fx(f, df, a)
        self.plot()
    
    def f_of_a(self, f, a):
        return f(a)
    
    def t_of_fx(self, f, df, a):
        fa = self.f_of_a(f, a) 
        
        dfa = df(a)
        
        return fa + dfa*(self.x - a)    
    
    def plot(self):
        plt.plot(self.x, self.f_of_x)
        plt.plot(self.x, self.lin_of_x_at_a)
        
        plt.legend(['f(x)', "T(f(x))"])
        
        plt.ylim(-2, 2)
        plt.xlim(-10,10)
        
        plt.show() #
```

### Let's linearize sin(x)

```python
def f(x):
    return np.sin(x) 

def df(x):
    return np.cos(x)

init_x = np.linspace(-10,10,100) # 100 linearly spaced numbers
l = Linearizer(f, df, init_x, 5)
```

### Linearize arctan(x)

```python
def f(x):
    return np.arctan(x)

def df(x):
    return 1/ (1 + x**2)

init_x = np.linspace(-10,10,100) # 100 linearly spaced numbers
l = Linearizer(f, df, init_x, 0)
```
