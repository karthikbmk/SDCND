### Linearization using Taylor series:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
class Linearizer:
    def __init__(self, f, df, init_x, a):
        self.x = init_x
        self.f_of_x = self.f_of_a(f, self.x)
        self.lin_of_x_at_a = self.t_of_fx(f, df, a)
        self.plot()
    
    def f_of_a(self, f, a):
        return f(a)
    
    def t_of_fx(self, f, df, a):
        fa = self.f_of_a(f, a) 
        
        dfa = df(a)
        
        return fa + dfa*(self.x - a)    
    
    def plot(self):
        plt.plot(self.x, self.f_of_x)
        plt.plot(self.x, self.lin_of_x_at_a)
        plt.legend(['f(x)', "T(f(x))"])
        plt.show() #
```

### Let's linearize sin(x)

```python
def f(x):
    return np.sin(x) 

def df(x):
    return np.cos(x)

init_x = np.linspace(-10,10,100) # 100 linearly spaced numbers
l = Linearizer(f, df, init_x, 5)
```

### Let's linearize the polynomial x^3 + x^2 + x

```python
def f(x):
    return x**3 + x**2 + x

def df(x):
    return 3*x**2 + 2*x + 1

init_x = np.linspace(-10,10,100) # 100 linearly spaced numbers
l = Linearizer(f, df, init_x, 5)
```
